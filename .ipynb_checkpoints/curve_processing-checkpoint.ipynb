# Curve processing

This notebook defines classes and functions for reading, interpolating, and formating all the melting curves prior to feeding them into ML pipelines.

It also defines an augmentation class that applies random data transformations that do not alter the prediction targets

```python
# import all the neccessary modules
import numpy as np
import pandas as pd
import json
from math import ceil
from scipy.signal import savgol_filter
from sklearn.utils import shuffle
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline

from IPython.display import display
```

```python
# global variables
CLF_TARGET = "clf_target"
REG_TARGET = "reg_target"
ID_COL = "id"
TEMP_COL = "temp"
SIG_COL = "signal"
DATA_FILE = "benchmark_set.json"
```

## Read in data file (JSON)

```python
# load data from the JSON file
with open(DATA_FILE, "r") as infile:
    data = json.load(infile)

# convert json string to DataFrame
curves_json = json.loads(data)
curves = pd.DataFrame(curves_json)
curves
```

## Process dataset

```python
# remove nans using linear interpolation
# then sample the data at fixed temp. again using linear interpolation
# finally smooth the data using a filter

def process_curve(row, scale=False, id_col=ID_COL, temp_col=TEMP_COL, sig_col=SIG_COL, clf_col=CLF_TARGET,
                  reg_col=REG_TARGET, start_temp=20, end_temp=100, sampling=100, smoothing=(9,2)):
    """produces re-sampled, interpolated, filtered, and re-scaled fluorescence curves
    smoothing - tuple of inegers (window_size, polynomial_power): parameters of the Savitzky-Golay
    smoothing algorithm"""
    # define data sampling points
    sampl = np.linspace(start_temp, end_temp, num=sampling)
    temp, fl = np.array(row[temp_col]), np.array(row[sig_col])
    # remove NaN in fl with interpolated values using linear interp.
    notNanVals = np.logical_not(np.isnan(fl)) # define a non Nan filter
    xp = notNanVals.nonzero()[0] # get indexes of non NaN elements
    fp = fl[notNanVals] # index and extract the fluorescence values that are not NaN
    x  = np.isnan(fl).nonzero()[0] # get indexes of NaN elements in fl
    fl[np.isnan(fl)] = np.interp(x, xp, fp) # replace NaNs in fl with interpolated values
    #plt.plot(temp, fl)
    # re-sample the data at sampl points using linear interp.
    resampled = np.interp(sampl, temp, fl)
    #plt.plot(sampl, resampled)
    # smooth the curve using a Savitzky-Golay filter (pol. interpolation)
    filtered = savgol_filter(resampled, smoothing[0], smoothing[1])
    # do internal scaling if the option is set (default)
    if scale:
        minVal, maxVal = np.amin(filtered), np.amax(filtered)
        filtered = (filtered - minVal) / (maxVal - minVal)
    ser = pd.concat([pd.Series(filtered), pd.Series(row[[id_col, clf_col, reg_col]])], axis=0)
    return ser

def process_dataset(dataset, scale=False, start_temp=20, end_temp=100, sampling=100, smoothing=(9,2),
                   id_col=ID_COL):
    """produces the final structured/processed dataset for ML"""
    processed = dataset.apply(process_curve, axis=1,
                              scale=scale, start_temp=start_temp, end_temp=end_temp, smoothing=smoothing)
    # the label column dtype is converted from int64 -> float64 by the merge operation
    # this is a pandas BUG! Workaround: convert dtype back to int64
    processed[id_col] = processed[id_col].astype(int)
    return processed
```

```python
processed = process_dataset(curves, scale=False)
processed.set_index(ID_COL, inplace=True)
processed
```

### Plot an example of processed data

```python
# plot an example of the processed dataset
def plot_processed_data(data, samples=20, seed=None, n_cols=4, clf_col=CLF_TARGET, reg_col=REG_TARGET):
    """displays the sample curves and their class labels"""
    # shuffle the dataset
    if seed:
        np.random.seed(seed=seed)
    curves_sample = data.sample(n=samples)
    targets = [clf_col, reg_col]
    data_cols = [c for c in curves_sample.columns if c not in targets]
    n_rows = ceil(samples / n_cols)
    # set the height of the plot
    height = n_rows * 3
    fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(20,height), sharex=True, sharey=True)
    # plot everything in the data array and annotate with the labels array
    for row, (idx, curve) in enumerate(curves_sample.iterrows()):
        fig_row = row // n_cols
        fig_col = row % n_cols
        axes[fig_row, fig_col].plot(curve[data_cols], label="class %i" %(curve[clf_col]))
        axes[fig_row, fig_col].axvline(curve[reg_col], color="r")
        axes[fig_row, fig_col].legend(loc="lower right")
    plt.show()
```

## Data Augmentation

Here I will try to use data augmentation techniques to increase the train/test set size and make the classifier/regressor more robust to expected variations in the data. The variations (and the data transformation methods I will use) are:

-  Scaling: multiply a curve by a small random number centered around 1 (+/- 0.1 * st.dev in train. set)(normal distr.)
-  Offset: add a small random number centered around 0 (+/- 0.1 * st.dev in train. set)(normal distr.)
-  slope: add a small kx term to each curve (x is the temp and k a small random number centered around 0: -0.05 - 0.05)(uniform distr.)
-  Flip vertically by 180 degrees (rotate around axis ax=mean(Fl)): ax = np.mean(Fl), new Fl = 2 * ax - Fl

```python
# define all the data transformations via a custom class

class Transformer(object):
    """a class for data augmentation operations"""
    def __init__(self, data, targets, random_seed=None, flip_prob=0.3, scale_factor=0.1,
                offset_factor=0.1, slope_factor=0.005):
        """data is a numpy array of shape (num_samples, n_features)
        targets is a numpy array of shape (num_samples, n_targets)"""
        self.data = np.array(data)
        self.targets = np.array(targets)
        self.rng = np.random.RandomState(seed=random_seed)
        self.stdev = self.calc_std()
        self.flip_prob = flip_prob
        self.scale_factor = scale_factor
        self.offset_factor = offset_factor
        self.slope_factor = slope_factor
        self.curve_size = data.shape[1]
    
    def calc_std(self):
        """removes outliers from the training dataset and computes the stdev
        on only the values that make sense"""
        cleaned = self.data[np.logical_and(self.data > 0, self.data < 5)]
        return np.std(cleaned)
        
    def rand_scale(self, curve):
        """scales the curve by a small random number"""
        return curve * self.rng.normal(loc=1, scale=self.scale_factor*self.stdev)
    
    def rand_offset(self, curve):
        """adds random offset to the curve"""
        return curve + self.rng.normal(loc=0, scale=self.offset_factor*self.stdev)
    
    def rand_slope(self, curve):
        """adds random slope to the curve"""
        return curve + self.rng.uniform(low=-self.slope_factor, high=self.slope_factor) * \
            np.linspace(0, self.curve_size, num=self.curve_size)
    
    def flip(self, curve):
        """flips the curve vertically around the y-mean with probability = self.flip_prob"""
        if self.rng.rand() < self.flip_prob:
            axis = np.mean(curve)
            return 2 * axis - curve
        else:
            return curve
    
    def transform_curve(self, curve):
        """apply transformations to a single curve"""
        operations = [self.rand_scale, self.rand_offset, self.rand_slope, self.flip]
        # randomly shuffle the operations and apply them one after the other
        self.rng.shuffle(operations)
        transformed = np.array(curve)
        for op in operations:
            transformed = op(transformed)
        return transformed

    def augment_data(self, expand=10):
        """expand dataset expand-fold by augmenting each curve expand-times"""
        expanded_data, expanded_targets = [], []
        for idx, curve in enumerate(self.data):
            for n in range(expand):
                expanded_data.append(self.transform_curve(curve))
                expanded_targets.append(self.targets[idx])
            
        return (np.array(expanded_data), np.array(expanded_targets))
```

### Visualize the augmentation results of a single curve

```python
# define a function for plotting all transformed variants of a curve
def plot_transform(data, targets, curve, expand=10, seed=None, **kwargs):
    """performs a set of transform operations expand times on an
    input curve and displayes the results on a single plot"""
    trsf = Transformer(data, targets, random_seed=seed, **kwargs)
    transf_curves = []
    for i in range(expand):
        transf_curves.append(trsf.transform_curve(curve))
    
    # plot all the augmented curves + original curve
    plt.figure(figsize=(10, 8))
    plt.plot(curve, label='original curve', color='deeppink', linestyle=':', linewidth=4)
    for i, c in enumerate(transf_curves):
        plt.plot(c, lw=2, label='transformed curve %i' %(i))
    
    plt.title('transformations of a melting curve')
    plt.legend(loc="lower right")
    plt.show()
```

```python
# test on a single curve
plot_transform(data.values, targets.values, data.loc[149,:].values, expand=5, random_seed=None)
```

### Augment the entire dataset

```python
# create the transformer instance and augment the whole dataset
trsf = Transformer(data.values, targets.values)
data_aug, targets_aug = trsf.augment_data(expand=5)
```

```python
# plot the augmented data to see if everything looks Ok
def plot_augmented_data(data, targets, n_cols=4, sample=20, seed=None):
    """displays all the curves and their class targets"""
    # create a random sample
    if seed:
        np.random.seed(seed)
    
    idxs = np.random.randint(0, data.shape[0] + 1, size=sample)
    data_sample, targets_sample = data[idxs,:], targets[idxs,:]
    n_rows = ceil(sample / n_cols)
    # set the height of the plot
    height = n_rows * 3
    fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(20,height), sharex=True, sharey=True)
    # plot everything in the data array and annotate with the targets array
    for idx, sample_curve in enumerate(data_sample):
        row = idx // n_cols
        col = idx % n_cols
        axes[row, col].plot(sample_curve, label="class %i" %(int(targets_sample[idx,0])))
        axes[row, col].axvline(targets_sample[idx,1], color="r")
        axes[row, col].legend(loc="lower right")
    plt.show()
```

```python
plot_augmented_data(data_aug, targets_aug)
```
