# Learning curve shape and predicting targets

Using several ML algorithms to do the following:
1. predict class label of each curve
2. if the class label is 1, predict temp. stability value (regression target)

## Import modules and data/targets

```python
# import the modules
import numpy as np
from scipy import interp
from itertools import cycle
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
%matplotlib inline

import os

from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor, export_graphviz
import graphviz
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.model_selection import train_test_split, cross_val_score, learning_curve, \
    validation_curve, GridSearchCV
from sklearn.metrics import classification_report, roc_curve, auc
```

```python
# load data and targets arrays
filename = "data_targets_aug.npz"
with np.load(filename) as npzfile:
    data = npzfile["data"]
    targets = npzfile["targets"]
```

```python
# split into classification and regression subsets
data_clf = data.copy()
targets_clf = targets[:,0].copy()

reg_mask = ~np.isnan(targets[:,1])
data_reg, targets_reg = data[reg_mask], targets[:,1][reg_mask]
```

## Classification of curves

### Class label definitions:
-  **0**: Noisy data, no information, failed experiment, discard
-  **1**: Data with high enough signal to obtain temp. stabilities

### Split classification data into a train/test set

```python
# split data for performance evaluation (use stratified sampling due to class imbalances)
data_clf_train, data_clf_test, labels_train, labels_test = train_test_split(data, targets_clf, test_size=0.2,
                                                                   random_state=0, stratify=targets_clf)
```

### Useful function defnitions

```python
# Plot a learning curve to see if there are issues with data size (also variance/bias issue diagnostics)
def plot_learn_curve(clf, data_train, labels_train):
    """plot a learning curve for the classifier clf using the training data
    data_train, class labels labels_train with 10-fold cross-validation"""
    train_szs, train_scs, test_scs = learning_curve(estimator=clf, X=data_train,
                                                   y=labels_train, train_sizes=np.linspace(0.1, 1.0, 20),
                                                    cv=10, n_jobs=-1)
    train_mean = np.mean(train_scs, axis=1) # array of means across cvs
    train_std = np.std(train_scs, axis=1) # array of std across cvs
    test_mean = np.mean(test_scs, axis=1)
    test_std = np.std(test_scs, axis=1)

    # make the plots
    plt.plot(train_szs, train_mean, color="blue", marker="o", markersize=5, label="training accuracy")
    plt.fill_between(train_szs, train_mean + train_std, train_mean - train_std, alpha=0.15, color="blue")
    plt.plot(train_szs, test_mean, color="green", linestyle="--", marker="s", markersize=5, 
             label="validation accuracy")
    plt.fill_between(train_szs, test_mean + test_std, test_mean - test_std, alpha=0.15, color="green")
    plt.grid()
    plt.legend(loc="lower right")
    plt.xlabel("training set size")
    plt.ylabel("accuracy")
    plt.show()
    

def plot_val_curve(clf, data_train, labels_train, param_name, param_range):
    """makes a validation plot - scan param_name and evaluate its effect on the classifier
    performance"""
    train_scs, test_scs = validation_curve(estimator=clf, X=data_train,
                                                      y=labels_train, param_name=param_name,
                                                      param_range=param_range, cv=10, n_jobs=-1)
    train_mean = np.mean(train_scs, axis=1) # array of means across cvs
    train_std = np.std(train_scs, axis=1) # array of std across cvs
    test_mean = np.mean(test_scs, axis=1)
    test_std = np.std(test_scs, axis=1)

    # make the plots
    plt.plot(param_range, train_mean, color="blue", marker="o", markersize=5, label="training accuracy")
    plt.fill_between(param_range, train_mean + train_std, train_mean - train_std, alpha=0.15, color="blue")
    plt.plot(param_range, test_mean, color="green", linestyle="--", marker="s", 
             markersize=5, label="validation accuracy")
    plt.fill_between(param_range, test_mean + test_std, test_mean - test_std, alpha=0.15, color="green")
    plt.grid()
    plt.legend(loc="lower right")
    plt.xlabel(param_name)
    plt.ylabel("accuracy")
    plt.show()

def plot_roc_curve(clf, data_test, labels_test, clf_model):
    """plots the various ROC curves with AUC.
    clf - a trained classifier instance"""
    # predict test set class membership probabilities/scores
    labels_prob = clf.predict_proba(data_test)
    
    # Compute ROC curve and AUC
    fpr, tpr, _ = roc_curve(labels_test, labels_prob[:,1])
    roc_auc = auc(fpr, tpr)

    # Plot the ROC curve
    lw = 2
    plt.figure(figsize=(10, 8))
    plt.plot(fpr, tpr, lw=lw, label='AUC = {0:0.2f}'.format(roc_auc))
    # plot diagonal
    plt.plot([0, 1], [0, 1], 'k--', lw=lw)
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('FPR')
    plt.ylabel('TPR')
    plt.title('ROC curve for {} classifier'.format(clf_model))
    plt.legend(loc="lower right")
    plt.show()
```

### Decision Trees

Let's first test how accurately decision trees are able to classify the data and then perform a grid search to find the optimal model hyperparameters and evaluate on the holdout set 

```python
clf_tree = DecisionTreeClassifier(criterion="gini", max_depth=10, class_weight="balanced")
```

#### DT learning curve 

```python
plot_learn_curve(clf_tree, data_clf_train, labels_train)
```

#### DT validation curve

```python
tree_depth_range = [d for d in range(10, 50, 5)]
plot_val_curve(clf_tree, data_clf_train, labels_train, "max_depth", tree_depth_range)
```

Seems like the best tree depth is around 25

#### Visualize the best tree as a graph

```python
# set optimal hyperparameters
clf_tree.set_params(max_depth=25)

# train on the entire training set and evaluate on the test set
clf_tree.fit(data_clf_train, labels_train)
print("*** Best tree accuracy: ", clf_tree.score(data_clf_test, labels_test))

# visualize as a graph
dot_data = export_graphviz(clf_tree, out_file=None, filled=True, rounded=True, rotate=True)
tree_graph = graphviz.Source(dot_data)
tree_graph
```

#### Plot ROC curve for the best DT

```python
plot_roc_curve(clf_tree, data_clf_test, labels_test, "Decision Tree")
```

### Random Forests

Explore the performance of RF classifiers and tune the model hyperparameters for optimal performance

```python
# create an ensamble instance
rfc = RandomForestClassifier(n_estimators=100, class_weight="balanced")
```

#### RF learning curve

```python
plot_learn_curve(rfc, data_clf_train, labels_train)
```

#### RF validation curve

```python
n_trees_range = [r for r in range(25, 500, 50)]
plot_val_curve(rfc, data_clf_train, labels_train, "n_estimators", n_trees_range)
```

#### Tune the RF hyperparameters via grid search

```python
param_grid = {"n_estimators": [d for d in range(25, 300, 75)],
             "max_features": [5, 10, 20, 40, 80]}

gs = GridSearchCV(estimator=rfc, param_grid=param_grid, cv=10, n_jobs=-1)
gs = gs.fit(data_clf_train, labels_train)
```

```python
# CV score of best estimator
gs.best_score_
```

```python
gs.best_estimator_
```

```python
# get the best estimator and train on the entire training set
rfc_best = gs.best_estimator_
rfc_best.fit(data_clf_train, labels_train)
```

#### Plot ROC curve for the best RF

```python
plot_roc_curve(rfc_best, data_clf_test, labels_test, "Random Forest")
```
